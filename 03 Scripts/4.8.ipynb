### 4. Follow the instructions in the Exercise for creating a loyalty flag for existing customers using the transform() and loc() functions.

```python
# Create loyalty flag using transform() and loc() 

df_merge_large['max_order'] = df_merge_large.groupby(['user_id'])['order_number'].transform(np.max)
```

### 6. The team now wants to target different types of spenders in their marketing campaigns. This can be achieved by looking at the prices of the items people are buying. Create a spending flag for each user based on the average price across all their orders using the following criteria:
### - If the mean of the prices of products purchased by a user is lower than 10, then flag them as a “Low spender.”
### - If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as a “High spender.”

```python
# Aggregate mean of the “prices_y” column grouped by “user_id” 

df_merge_large.groupby('user_id').agg({'prices_y': ['mean']})
```

```python
df_merge_large['Average_spending'] = df_merge_large.groupby(['user_id'])['prices_y'].transform(np.mean)
```

```python
# Create 'Low spender'

df_merge_large.loc[df_merge_large['Average_spending'] < 10, 'spending_flag'] = 'Low spender'
```

```python
# Create 'High spender'

df_merge_large.loc[df_merge_large['Average_spending'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Check frequency

df_merge_large['spending_flag'].value_counts(dropna = False)
```

### 7. In order to send relevant notifications to users within the app (for instance, asking users if they want to buy the same item again), the Instacart team wants you to determine frequent versus non-frequent customers. Create an order frequency flag that marks the regularity of a user’s ordering behavior according to the median in the “days_since_prior_order” column. The criteria for the flag should be as follows:
### - If the median of “days_since_prior_order” is higher than 20, then the customer should be labeled a “Non-frequent customer.”
### - If the median is higher than 10 and lower than or equal to 20, then the customer should be labeled a “Regular customer.”
### - If the median is lower than or equal to 10, then the customer should be labeled a “Frequent customer.”

```python
# Aggregate median of the “days_since_prior_order” column grouped by “user_id” 

df_merge_large.groupby('user_id').agg({'days_since_prior_order': ['median']})
```

```python
# Create 'frequency_flag'

df_merge_large['frequency_flag'] = df_merge_large.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# Rename 'frequency_flag' column to 'Median_days_since_prior_order'

df_merge_large.rename(columns = {'frequency_flag' : 'Median_days_since_prior_order'}, inplace = True)
```

```python
# Create 'Non-frequent customer'

df_merge_large.loc[df_merge_large['Median_days_since_prior_order'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
# Create 'Regular customer'

df_merge_large.loc[(df_merge_large['Median_days_since_prior_order'] <= 20) & (df_merge_large['Median_days_since_prior_order'] > 10), 'frequency_flag'] = 'Regular customer'
```

```python
# Create 'Frequent customer'

df_merge_large.loc[df_merge_large['Median_days_since_prior_order'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Check frequency

df_merge_large['frequency_flag'].value_counts(dropna = False)
```

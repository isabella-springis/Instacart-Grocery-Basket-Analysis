```python
# Define function for price_label

def price_label(row):

  if row['prices_x'] <= 5:
    return 'Low-range product'
  elif (row['prices_x'] > 5) and (row['prices_x'] <= 15):
    return 'Mid-range product'
  elif row['prices_x'] > 15:
    return 'High range'
  else: return 'Not enough data'
```
